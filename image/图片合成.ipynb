```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# 
img1 = mpimg.imread('model_architecture.png')

# 
img2 = mpimg.imread('model_flowchart.jpg')

# 
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# 
ax1.imshow(img1)
ax1.axis('off')  # 

# 
ax2.imshow(img2)
ax2.axis('off')  # 

# 
plt.tight_layout()

#  TIFF 
plt.savefig('combined_image.tif', dpi=300, format='tif')

#  PNG 
plt.savefig('combined_image.png', dpi=300, format='png')

# 
plt.show()
```
